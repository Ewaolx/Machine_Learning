```python
import pandas as pd
df =  pd.read_table("smsspamcollection/SMSSpamCollection",sep='\t',names=['y_target','x_message'])
df.head()
```

```python
def encode(x): return 1 if (x == 'spam') else 0
df['label']=df.y_target.apply(encode)
df.drop(['y_target'],axis=1)
```

```python
from sklearn.cross_validation import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df['x_message'], 
                                                    df['label'], 
                                                    random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

count_vector = CountVectorizer()
xtrain=count_vector.fit_transform(x_train)
print(count_vector.vocabulary_.get("car"))
xtest=count_vector.transform(x_test)
xtest.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(xtrain,y_train)
```

```python

from sklearn.metrics import accuracy_score


prediction=nb.predict(xtest)
acc=accuracy_score(prediction,y_test)
print("Naive_Bayes__Accuracy_Score:",acc)
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
l.fit(xtrain,y_train)
lp=l.predict(xtest)
ap=accuracy_score(lp,y_test)
print("Logisitic_Regression_Accuracy_Score:",ap)
```

```python

from sklearn.metrics import r2_score
print('Naive_Bayes_RMS_Value:',r2_score(y_test,prediction))
print('Logisitic_Regression_RMS_Value:',r2_score(y_test,lp))
```
